```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
```

```python
path = "../input/pickle/"

test_book = pd.read_pickle(path + "test_book.zip")
test_game = pd.read_pickle(path + "test_game.zip")
test_goods = pd.read_pickle(path + "test_goods.zip")
test_move = pd.read_pickle(path + "test_move.zip")
test_music = pd.read_pickle(path + "test_music.zip")
test_pc_game = pd.read_pickle(path + "test_pc_game.zip")
```

```python
test_book
```

```python
test = pd.concat([test_book, test_game])
test = pd.concat([test, test_goods])
test = pd.concat([test, test_move])
test = pd.concat([test, test_music])
test = pd.concat([test, test_pc_game])
```

```python
test = test.sort_index()

path = "../submission/"

submission = pd.read_csv(path + "sample_submission.csv", header=None)

submission.iloc[:,[1]] = test["y"].values
submission.head()
```

```python
submission.to_csv(path + "submission.csv", index=False, header=None)
```
